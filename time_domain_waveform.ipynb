### Display the waveform of time-domain of each category

```python
#sample one instance from every category
file_names_list=[]
common_name_list=[]
n=len(metadata)
for i in range(n):
    if metadata.iloc[i][2] not in common_name_list:
        common_name_list.append(metadata.iloc[i][2])
        file_names_list.append(metadata.iloc[i][0])
#print(file_names_list)
#print(common_name_list)
```

```python
import IPython.display as ipd
import librosa
import librosa.display
import matplotlib.pyplot as plt
plt.rcParams.update({'figure.max_open_warning': 0})
size_file_names_list=len(file_names_list)
for i in range(size_file_names_list):
    path='../data/'+file_names_list[i]
    plt.figure(figsize=(12,4))
    plt.title(common_name_list[i])
    data,sample_rate = librosa.load(path)
    _ = librosa.display.waveshow(data,sr=sample_rate)
    ipd.Audio(path)
```
